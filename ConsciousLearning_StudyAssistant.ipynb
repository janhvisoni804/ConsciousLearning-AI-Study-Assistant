```python
from IPython.display import Audio, display
import google.generativeai as genai
import os

#insert your key
import getpass, os
os.environ["GEMINI_API_KEY"] = getpass.getpass("Enter your Gemini API Key: ")

# âœ¨ Configure Model
genai.configure(api_key=os.environ["GEMINI_API_KEY"])
model = genai.GenerativeModel("models/gemini-2.5-flash")

# ğŸ§ Text â†’ Speech
def speak(text):
    from gtts import gTTS
    tts = gTTS(text)
    file = "voice.mp3"
    tts.save(file)
    return Audio(file, autoplay=True)

# ğŸ§  AI Helper
def ask_ai(msg):
    reply = model.generate_content([msg]).text
    return reply

# ğŸ’¾ Student Memory
profile = {}

# ğŸ¤– Study Agents
def menu():
    print("""
========== ConsciousLearning ==========
1ï¸âƒ£ Study Plan
2ï¸âƒ£ Quiz Generator
3ï¸âƒ£ Explain Topic
4ï¸âƒ£ Summarize Notes
5ï¸âƒ£ Exit
=======================================
""")
    choice = input("Choose (1-5): ")

    if choice == "1": return study_plan()
    elif choice == "2": return quiz()
    elif choice == "3": return explain()
    elif choice == "4": return summary()
    else:
        print("\nGoodbye! Keep learning ğŸŒŸ")
        display(speak("Goodbye! Keep learning."))
        return

def start():
    print("ğŸ‘‹ Welcome to ConsciousLearning!")
    print("Which level?\n1 School  2 College  3 Competitive Exams")
    profile["level"] = input("â¡ ")

    print("\nEnter subjects comma separated:")
    profile["subjects"] = [s.strip() for s in input("â¡ ").split(",")]

    print("\nAll Set! ğŸš€")
    menu()

def study_plan():
    subjects = ", ".join(profile["subjects"])
    ans = ask_ai(f"Create weekly study plan for {subjects} in simple English.")
    print("\nğŸ“š Study Plan:\n", ans)
    display(speak("Here is your study plan."))
    return menu()

def quiz():
    sub = input("\nQuiz for which subject? â¡ ")
    ans = ask_ai(f"Generate 5 MCQs with answers for {sub}.")
    print("\nğŸ“ Quiz:\n", ans)
    display(speak("Quiz ready!"))
    return menu()

def explain():
    topic = input("\nWhich topic to explain? â¡ ")
    ans = ask_ai(f"Explain {topic} in simplest form with examples.")
    print("\nâœ¨ Explanation:\n", ans)
    display(speak("Explained the topic."))
    return menu()

def summary():
    print("\nPaste your notes to summarize:")
    text = input("â¡ ")
    ans = ask_ai(f"Summarize:\n{text}")
    print("\nğŸ§¾ Summary:\n", ans)
    display(speak("Your summary is ready."))
    return menu()

# ğŸš€ Start App
start()
```
